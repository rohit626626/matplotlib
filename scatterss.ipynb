```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("data.csv")
df
```

```python
x=df['Study hrs']
y=df['Percantege']
```

```python
plt.scatter(x,y)
plt.title('student info')
plt.xlabel('study hrs')
plt.ylabel('percantage')
plt.show()
```

```python
plt.scatter(x,y,c='r', marker='*', s=100,alpha=0.5, linewidths=10,edgecolors='g')
plt.title('student info')
plt.xlabel('study hrs')
plt.ylabel('percantage')
plt.show()
```

```python
plt.scatter(x,y,c='r', marker='*', s=100,alpha=0.5, linewidths=10,edgecolors='g')
plt.scatter(x, df['Class'], c='y', marker='o', s=100,alpha=0.5, linewidths=10,edgecolors='y')
plt.title('student info')
plt.xlabel('study hrs')
plt.ylabel('percantage and class')
plt.show()
```
