```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age=np.random.randint(18,45,100)
py_students_age=np.random.randint(15,40,100)
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
plt.hist(ml_students_age)
plt.title("Student histograms")
plt.xlabel("Student age")
plt.ylabel("No. Student age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins)
plt.title("Student histograms")
plt.xlabel("Student age")
plt.ylabel("No. Student age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8)
plt.title("Student histograms")
plt.xlabel("Student age")
plt.ylabel("No. Student age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8, histtype='bar', orientation="vertical", color='r', label='Ml Student')
plt.title("Student histograms")
plt.xlabel("Student age")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8, histtype='bar', orientation="vertical", color='r', label='Ml Student')
plt.hist(py_students_age, bins, rwidth=0.8, histtype='bar', orientation="vertical", color='g', label='Py Student')
plt.title("Student histograms")
plt.xlabel("Student age")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([ml_students_age,py_students_age], bins, rwidth=0.8, histtype='bar', orientation="vertical", color=['g','r'], label=['Ml Student', 'PY Student'])
plt.title("Student histograms")
plt.xlabel("Student age")
plt.ylabel("No. Student age")
plt.legend()
plt.show()
```
