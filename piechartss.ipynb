```python
import matplotlib.pyplot as plt
```

```python
classes=['Python','R','Machine Learning', 'AI','Data Science']
class1_students=[45,15,35,25,30]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_students, labels=classes)
plt.show()
```

```python
explode=[0.03,0,0.1,0,0]
colors=['c','b','r','b','y']
textprops={'fontsize':15}
plt.pie(class1_students, labels=classes, explode=explode, colors=colors, autopct='%0.2f%%', shadow=True, radius=1.4, startangle=90, textprops=textprops)
plt.show()
```

```python
explode=[0.03,0,0.1,0,0]
colors=['c','b','r','b','y']
textprops={'fontsize':15}
plt.pie(class1_students, labels=classes, explode=explode, colors=colors, autopct='%0.2f%%', shadow=True, radius=1.4, startangle=90, textprops=textprops)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,6))
wedgeprops={"linewidth":4, 'width':1, 'edgecolor':'k'}
plt.pie(class1_students, labels=classes, explode=explode, colors=colors, autopct='%0.2f%%', pctdistance=0.6, shadow=True, labeldistance=1.6, startangle=270, radius=1.1, counterclock=True, wedgeprops=wedgeprops, textprops=textprops, center=(2,3),frame=True, rotatelabels=True)
plt.show()
```
