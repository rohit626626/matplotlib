```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=["Python", "R", "AI", "ML", "DS"]
class1_studnets=[30,10,20,25,10]
class2_students=[40,5,20,20,10]
class3_students=[30,5,30,15,15]
```

```python
plt.bar(classes, class1_studnets)
```

```python
plt.barh(classes, class1_studnets)
```

```python
plt.bar(classes, class1_studnets, width=0.2, align="edge", color='g',edgecolor='m', linewidth=5, alpha=0.8, linestyle='--',label="Class 1 Student")
```

```python
plt.bar(classes, class1_studnets, width=0.2, align="edge", color='g',edgecolor='m', linewidth=5, alpha=0.8, linestyle='--',label="Class 1 Student")
plt.title("Bar Chart of Classes", fontsize=18)
plt.xlabel("Classes", fontsize=15)
plt.ylabel("No of Students", fontsize=15)
plt.show()
```

```python
plt.bar(classes, class1_studnets, width=0.2,  color='g',label="Class 1 Student")
plt.bar(classes, class2_students, width=0.2,  color='b',label="Class 2 Student")
plt.title("Bar Chart of Classes", fontsize=18)
plt.xlabel("Classes", fontsize=15)
plt.ylabel("No of Students", fontsize=15)
plt.show()
```

```python
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index, class1_studnets, width,  color='g',label="Class 1 Student")
plt.bar(classes_index+width, class2_students, width,  color='b',label="Class 2 Student")
plt.bar(classes_index+width+width, class2_students, width,  color='g',label="Class 3 Student")
plt.xticks(classes_index+width, classes, rotation=20)
plt.title("Bar Chart of Classes", fontsize=18)
plt.xlabel("Classes", fontsize=15)
plt.ylabel("No of Students", fontsize=15)
plt.legend()
plt.show()
```

```python
classes_index=np.arange(len(classes))
width=0.2
plt.barh(classes_index, class1_studnets, width,  color='g',label="Class 1 Student")
plt.barh(classes_index+width, class2_students, width,  color='b',label="Class 2 Student")
plt.barh(classes_index+width+width, class2_students, width,  color='g',label="Class 3 Student")
plt.yticks(classes_index+width, classes, rotation=20)
plt.title("Bar Chart of Classes", fontsize=18)
plt.ylabel("Classes", fontsize=15)
plt.xlabel("No of Students", fontsize=15)
plt.legend()
plt.show()
```
